```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
file_path = "Files/sales_transactions.csv"
```

```python
df_bronze = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
```

```python
display(df_bronze)
```

```python
df_bronze.printSchema()
```

```python
print("Total Records:", df_bronze.count())
```

```python
df_bronze = df_bronze.withColumn(
    "ingestion_timestamp",
    current_timestamp()
)
```

```python
display(df_bronze)
```

```python
spark.sql("DROP TABLE IF EXISTS bronze_sales")
```

```python
df_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_sales")
```

```python
spark.sql("""
SELECT *
FROM bronze_sales
LIMIT 10
""").show()
```
